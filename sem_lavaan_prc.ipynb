```r
# measurement model
model1 = '
    latent =~ x1 + x2 + x3

'
# fit the measurement model using cfa()
model1_fit = cfa(model1, data = holz)
# get the summary 
summary_model1 = summary(model1_fit, fit.measures = T, 
                        standardized = T)
```

```r
3 * 4/2 # number of unique variances and covariance
# number of free parameters to estimate = 6
# 3 path coefficients from latent variable to manifest variables
# 3 path coefficients from the errors to manifest variables
# the latent variable variance is fixed at 1.00
# otherwise, the path coefficient from latent variable to the 
# first manifest variable is fixed at 1.0
# this frees the variance of the latent variable
```

```r
holz2 = HolzingerSwineford1939[, c(7:12)] # create a six variable data
head(holz2)
```

```r
# set up a structural model where latent variable 2 will
# regress on latent variable 1

model2 = '
    lat1 =~ x1 + x2 + x3
    lat2 =~ x4 + x5 + x6
    lat2 ~ lat1

'
model2_fit = sem(model2, data = holz2) # fit the model

summary_model2 = summary(model2_fit, standardized = T, 
                        fit.measures = T) # get the summary
```

```r
# get the modification indices
mi = modindices(model2_fit)
mi
```

```r
model3 = '
    lat1 =~ x1 + x2 + x3
    lat2 =~ x1 + x4 + x5 + x6
    lat2 ~ lat1

'
model3_fit = sem(model3, data = holz2) # fit the model

summary_model3 = summary(model3_fit, standardized = T, 
                        fit.measures = T) # get the summary
```

```r
model4 = '
    lat1 =~ x1 + x2 + x3
    lat2 =~ x1 + x2 + x4 + x5 + x6
    lat2 ~ lat1

'
model4_fit = sem(model4, data = holz2) # fit the model

summary_model4 = summary(model4_fit, standardized = T, 
                        fit.measures = T) # get the summary
```

```r
bjcov = lav_matrix_lower2full(c(648.07, 30.05, 8.64,
                               140.18, 25.57, 233.21 ))
colnames(bjcov) = rownames(bjcov) = c("salary",
                                     "school",
                                     "iq")
# specify the model
bjmodel = '
    salary ~ a*school + c*iq
    school ~ b*iq
    ind:= b*c
'
# fit model
bjfit = sem(bjmodel, sample.cov = bjcov, sample.nobs = 300)
```

```r
summary(bjfit)
```

```r
# latent growthe curve model
# measure over several time points
# Study change over time
# model trajectory of growth as linear or quadratic
# Random effects as latent variables
# Randome effects = intercepts and slopes (i and s)
# intercept = compared to baseline, what would be the
# the value at timepoint t and how does it grow/shrink
# slope = at what rate does it shrink/grow
# linear growth model, 4 timepoints t1 ... t4
# fixed coefficients for intercept and slope

growth_model = '
        i =~ 1*t1 + 1 *t2 + 1*t3 + 1*t4 # flat intercept
        s =~ 0*t1 + 1 * t2 + 2 * t3 + 3 * t4 # slope jumps by one pt
'
growth_fit = growth(growth_model,
                   data = Demo.growth) # fit the model
summary(growth_fit)
```

```r
# growth curve model with covariates
growth_model2 = '
        i =~ 1*t1 + 1 *t2 + 1*t3 + 1*t4 # flat intercept
        s =~ 0*t1 + 1 * t2 + 2 * t3 + 3 * t4 # slope jumps by one pt
        i ~ x1 + x2 # time fixed covariates 
        s ~ x1 + x2
        t1 ~ c1
        t2 ~ c2 
        t3 ~ c3
        t4 ~ c4


'
growth_fit2 = growth(growth_model2,
                   data = Demo.growth) # fit the model
summary(growth_fit2)
```
